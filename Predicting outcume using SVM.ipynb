```python
#import libraries
from sklearn import svm
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
#display the 10 first rows
data.head(10)
```

```python
#transform the Gender column to binary
confirme=LabelEncoder()
data['Gender']=confirme.fit_transform(data['Gender'])
data.head()
```

```python
#classify the index
bins=(-1,0,1,2,3,4,5)
health=['malnourished','underweight','fit','slightly overweight','overweight','extremly overweight']
data['Index']=pd.cut(data['Index'],bins=bins,labels=health)
```

```python
# drop the column index from the dataset 
X=data.drop('Index', axis=1) #training data
y=data['Index'] # test data
```

```python
#split the dataset
X_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
X_train.shape
```

```python
#use the DtandardScaler
s=StandardScaler()
X_train=s.fit_transform(X_train)
x_test=s.transform(x_test)
```

```python
#apply the SVM algorithm and fit and predict the model
clf =svm.SVC()
clf.fit(X_train,y_train)
pred=clf.predict(x_test)
print(classification_report(y_test,pred)) #accuracy is very good 85%
```

```python
print(confusion_matrix(y_test,pred)) #the confusion matrix
```
